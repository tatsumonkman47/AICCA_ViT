```python
model_dirs_output = ["ViT_1M_3layer",
                     "ViT_1M_1layers_64patches",
                     "ViT_1M_2layer_256patches",
                     "ViT_1M_2layers_64patches",
                     "ViT_1M_3layer_64patches",
                    ]
save_matrix = {}

for output_dir in model_dirs_output:
    versions = os.listdir(f"./{output_dir}/lightning_logs/")
    save_matrix[output_dir]=[pd.read_csv(f"./{output_dir}/lightning_logs/{version}/metrics.csv") for version in versions]
    

    
    
```

```python
val_mask = np.isfinite(save_matrix["ViT_1M_1layers_64patches"][0].val_acc)
val_step_masked = save_matrix["ViT_1M_1layers_64patches"][0].step[val_mask]
```

```python
plt.figure(figsize=(20,10))

plt.plot(save_matrix["ViT_1M_1layers_64patches"][0].step,save_matrix["ViT_1M_1layers_64patches"][0].train_acc,label="ViT_1M_1layer 64 patches train_acc",color="steelblue",alpha=0.7)
plt.plot(save_matrix["ViT_1M_2layer_256patches"][0].step,save_matrix["ViT_1M_2layer_256patches"][0].train_acc,label="ViT_1M_2layer 256 patches train_acc",color="indianred",alpha=0.7)


val_mask = np.isfinite(save_matrix["ViT_1M_2layer_256patches"][0].val_acc)
val_step_masked = save_matrix["ViT_1M_2layer_256patches"][0].step[val_mask]
val_acc_masked = save_matrix["ViT_1M_2layer_256patches"][0].val_acc[val_mask]
plt.plot(val_step_masked,val_acc_masked,color="indianred",linestyle="--",linewidth=4)
plt.scatter(val_step_masked,val_acc_masked,color="k",s=250,zorder=8)
plt.scatter(val_step_masked,val_acc_masked,label="ViT_1M_2layer_256patches val_acc",color="indianred",s=150,zorder=9)

val_mask = np.isfinite(save_matrix["ViT_1M_1layers_64patches"][0].val_acc)
val_step_masked = save_matrix["ViT_1M_1layers_64patches"][0].step[val_mask]
val_acc_masked = save_matrix["ViT_1M_1layers_64patches"][0].val_acc[val_mask]
plt.plot(val_step_masked,val_acc_masked,color="steelblue",linestyle="--",linewidth=4)
plt.scatter(val_step_masked,val_acc_masked,color="k",s=250,zorder=10)
plt.scatter(val_step_masked,val_acc_masked,label="ViT_1M_1layer val_acc",color="steelblue",s=150,zorder=11)


plt.legend()

plt.xlabel("Step",fontsize=25)
plt.xlabel("Accuracy",fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
    
```
